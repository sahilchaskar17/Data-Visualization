# TASK #2: PLOT INTERACTIVE BUBBLE CHART (SCATTERPLOT WITH SIZE)

```python
# Let's add a fourth variable "SOP" as the size 
fig = px.scatter(admission_df, x = 'GRE Score', y = 'Chance of Admit', color = 'University Rating', size = 'SOP') # size denotes the size of bubble
fig.show()
```

```python
# You can also add more data on hover using hover_data
fig = px.scatter(admission_df, x = 'GRE Score', y = 'Chance of Admit', color = 'University Rating', size = 'SOP', hover_data = ['LOR']) # size denotes the size of bubble
fig.show()
```

Extra Exploration
- **Modify the SOP column to make the bubble size variations more prominent** 

```python
admission_df['SOP'] = admission_df['SOP'] ** 2
fig = px.scatter(admission_df, x = 'GRE Score', y = 'Chance of Admit', color = 'University Rating', size = 'SOP', hover_data = ['LOR']) # size denotes the size of bubble
fig.show()
```

# TASK #4: PLOT INTERACTIVE MULTIPLE LINE PLOTS USING PLOTLY EXPRESS

```python
fig  = px.line()

for i in crypto_df.columns[1:]:
    fig.add_scatter(x = crypto_df['Date'], y = crypto_df[i], name = i)
    
fig.show()
```

# TASK #5. PLOT INTERACTIVE PIE CHARTS

```python
# Define a dictionary with all crypto allocation in a portfolio
# Note that total summation = 100% 
my_dict = {'allocation %': [20, 20, 20, 20, 20]}
my_dict
```

```python
crypto_df1 = pd.DataFrame(data = my_dict, index = ['BTC', 'ETH', 'LTC', 'XRP', 'ADA'])
crypto_df1
```

```python
# Use Plotly Express to plot a pie chart 
fig = px.pie(crypto_df1, values = 'allocation %', names = ['BTC', 'ETH', 'LTC', 'XRP', 'ADA'], title = 'Crypto Allocation')
fig.show()
```

```python
# adding hole to pie chart
fig = px.pie(crypto_df1, values = 'allocation %', names = ['BTC', 'ETH', 'LTC', 'XRP', 'ADA'], hole = 0.2, title = 'Crypto Allocation')
fig.show()
```

# TASK #6: PLOT INTERACTIVE BAR CHART 

```python
# Gapminder combines data from multiple sources in a time-series format
# Check this out: https://www.gapminder.org/data/

# data can be read directly with plotly if firewall or something else is opening the csv on excel
data = px.data.gapminder()
data
```

```python
# Gapminder combines data from multiple sources in a time-series format
# Check this out: https://www.gapminder.org/data/
# You can read the data directly as follows: data = px.data.gapminder()

# Alternatively, you can import the data as follows:
data = pd.read_csv('data/gapminder.csv')
data
```

```python
# Filter the data based on the country of choice
canada_df = data[data['country'] == 'Canada']
canada_df
```

```python
fig = px.bar(canada_df, x = 'year', y = 'pop', labels = {'pop': 'Population of Canada'}, height = 500)
fig.show()
```

```python
# You can add hoverdata and color (third dimension) as follows:
fig = px.bar(canada_df, x = 'year', y = 'pop', color = 'lifeExp', hover_data = ['gdpPercap'], labels = {'pop': 'Population of Canada'}, height = 500)
fig.show()
```

# TASK #7: PLOT INTERACTIVE GANTT CHART

```python
# Define Job #1 
job_1 = {'Task': 'Develop Content', 'Start': '2020-10-10', 'Finish': '2021-01-01'}
job_1
```

```python
# Define Job #2 
job_2 = {'Task': 'Film Videos', 'Start': '2021-01-01', 'Finish': '2021-03-03'}
job_2
```

```python
# Define Job #3
job_3 = {'Task': 'Edit Videos', 'Start': '2021-03-06', 'Finish': '2021-04-04'}
job_3
```

```python
project_df = pd.DataFrame([job_1, job_2, job_3])
project_df
```

```python
fig = px.timeline(project_df, x_start = 'Start', x_end = 'Finish', y = 'Task')
fig.update_yaxes(autorange = 'reversed')
fig.show()
```

Extra Exploration
- **Add an additional task of "Send the course for approval" that lasts for 1 day and that starts immediately after "Edit Videos & Upload Content"** 

```python
job_4 = {'Task': 'Send the course for approval', 'Start': '2021-04-04', 'Finish': '2021-04-05'}
project_df = pd.DataFrame([job_1, job_2, job_3, job_4])

fig = px.timeline(project_df, x_start = 'Start', x_end = 'Finish', y = 'Task')
fig.update_yaxes(autorange = 'reversed')
fig.show()
```

# TASK #8: PLOT INTERACTIVE SUNBURST

```python
# A sunburst plot represents hierarchial data as sectors laid out over several levels of concentric rings
restaurant_df = pd.read_csv('data/restaurant_mini.csv')
restaurant_df.dropna()
restaurant_df
```

```python
fig = px.sunburst(restaurant_df, path = ['Dining or Takeout', 'Day', 'Age'], values = 'Invoice')
fig.show()
```
